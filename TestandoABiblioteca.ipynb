-> Antes de iniciar os trabalhos do python com o mysql é necessário instalar 
a biblioteca PyMySQL usando a biblioteca pip

-> PyMySQL: O PyMySQL é uma biblioteca que funciona como um tradutor (ou
uma ponte) entre o nosso código python e o servidor de banco de dados 
MySQL. O PyMySQL é um driver de banco de dados (ou conector). Ele
implementa o protocolo de comunicação MySQL usando apenas código Python
puro. Isso é importante porque:

-> 1. Não precisa de dependências de sistemas: Diferente de outras
bibliotecas mais antigas que exigiam a instalação de compiladores
C ou arquivos do sistema operacional, o pymysql é fácil de instalar
(pip install PyMySQL) e roda em qualquer lugar que tenha python.

-> 2. Segue padrões: Ela segue a especificação PEP 249, que é
o padrão oficial do Python para banco de dados. Isso significa 
que se você aprender a usar o pymysql, saberá usar quase qualquer
outra biblioteca de banco de dados (como o PostgreSQL ou SQLite),
pois os comandos (connect, cursor, execute) são quase idênticos.


Testando se a biblioteca está funcionando: Vamos criar um trecho que irá utilizar o comando "SHOW DATABASES" do MySQL utilizando a biblioteca pymysql

```python
# Import da biblioteca pymysql
import pymysql

# connect: Função que tem como objetivo conectar o python no banco de 
# dados mysql. A função recebe como argumento:

    # -> host: Endereço do servidor (local que armazena os banco de dados).
    # -> user: Nome do administrador do banco de dados.
    # -> password: Senha para acessar o banco de dados.
    # -> database: banco de dados que queremos utilizar
conexao = pymysql.connect(
    
    host ='localhost',
    
    user = 'root',
    
    password = ''
    
    # Observação: Como nesse caso não estamos especificando a base de
    # de dados que queremos utilizar, o pymysql irá acessar todos os
    # bancos existentes no nosso mysql. 
)

# Tem como objetivo fazer a "ponte" entre o mysql e o python. Basicamente
# o cursor será responsável por enviar para o mysql os comandos realizados
# no código python. Além de transportar comandos, o cursor tem uma função
# vital: Memória de Posição. Imagine que pedimos ao banco de dados uma
# lista com 1 milhão de clientes:

        # -> O banco não joga 1 milhão de nomes de uma vez (isso travaria
        # nosso computador).
        
        # -> O cursor funciona como um dedo apontando para uma linha
        # de uma planilha.
        
        # -> Ele pega a primeira linha. Quando você pede a próxima, ele 
        # move o "dedo" para a linha de baixo. É por isso que conseguimos
        # fazer o for x in cursor: o cursor sabe exatamente em qual linha
        # ele está e qual é a próxima.
cursor = conexao.cursor()

# Execute: Tem como objetivo executar comandos mysql.
cursor.execute("SHOW DATABASES")

# Ira percorrer o cursor com a função execute que irá mostrar todos
# os bancos de dados presentes no nosso servidor 
for x in cursor:
    
    print(x)
```
